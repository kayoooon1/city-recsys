```python
import pandas as pd
```

```python
temp_df = pd.read_csv('all-data-datalit/mean_sd_temperature_cities.csv')
temp_df = temp_df.drop(['Unnamed: 0','Ref.'], axis=1)
temp_df = temp_df.rename(columns={'Country':'country'})
print(temp_df.shape)
temp_df.head()
```

```python
safety_df = pd.read_csv('all-data-datalit/safety-city-code.csv')
safety_df = safety_df.drop(['Unnamed: 0','code'], axis=1)
safety_df = safety_df.rename(columns={'Rank':'rank_safety','Safety Index':'safety_index'})
print(safety_df.shape)
safety_df.head()
```

```python
pollution_df = pd.read_csv('all-data-datalit/pollution-city-code.csv', sep=';')
pollution_df = pollution_df.drop(['Unnamed: 3'], axis=1)
#print(pollution_df.columns)
pollution_df = pollution_df.rename(columns={'City ':'city_ascii','Country ':'country','Index':'pollution_index'})
print(pollution_df.shape)
pollution_df.head()
```

```python
cost_living_df = pd.read_csv('all-data-datalit/cost-of-living-city-code.csv')
cost_living_df = cost_living_df.drop(['Unnamed: 0','code'], axis=1)
cost_living_df = cost_living_df.rename(columns={'Rank':'rank_cost_living','Cost of Living Plus Rent Index':'cost_live_rent_index',
                                               'Local Purchasing Power Index':'purchase_pow_index'})
print(cost_living_df.shape)
cost_living_df.head()
```

```python
traffic_df = pd.read_csv('all-data-datalit/trafficmerged.csv')
print(traffic_df.columns)
traffic_df = traffic_df.drop(['COUNTRY', 'Inefficiency.Index','Time.Exp..Index','Code'], axis=1)
traffic_df = traffic_df.rename(columns={'City':'city_ascii','Rank':'rank_traffic','country':'country','Time.Index..in.minutes.':'trffic_min_index'})
print(traffic_df.shape)
traffic_df.head()
```

```python
print("1",cost_living_df.shape,temp_df.shape)
final_df = pd.merge(cost_living_df, temp_df, how='inner',left_on='city_ascii',right_on='city_ascii',suffixes=('_col', '_temp'))
print("2",final_df.shape)
print("3",safety_df.shape)
#final_df.head()
final_df = pd.merge(final_df, safety_df, how='inner', left_on='city_ascii',right_on='city_ascii',suffixes=('_x', '_safe'))
print("4",final_df.shape)
print("5",pollution_df.shape)
final_df = pd.merge(final_df, pollution_df, how='inner', left_on='city_ascii',right_on='city_ascii',suffixes=('_x', '_pollutn'))
print("6",final_df.shape)
print("7",traffic_df.shape)
final_df = pd.merge(final_df, traffic_df, how='inner', left_on='city_ascii',right_on='city_ascii',suffixes=('_x', 'trffc'))
print("8",final_df.shape)
```

```python
print(final_df.columns)
final_df = final_df.drop(['country_col','country_x','country_pollutn','Country','rank_cost_living',
                         'rank_safety','rank_traffic'], axis=1)
final_df = final_df.rename(columns={'country_temp':'country'})
final_df = final_df.drop_duplicates(['city_ascii'])
print(final_df.shape)

final_df.head()
```

```python
lat_long_df = pd.read_csv('all-data-datalit/worldcities.csv')
lat_long_df = lat_long_df.drop(['city','iso2','iso3','admin_name','capital','population','id'], axis=1)
lat_long_df.head()
```

```python
final_df = pd.merge(final_df, lat_long_df, how='left', left_on=['city_ascii','country'],
                    right_on=['city_ascii','country'],suffixes=('_x', '_lat_long'))
final_df = final_df.drop_duplicates(['city_ascii'])
final_df.loc[(final_df['city_ascii']== 'Hong Kong'), 'lat'] = 22.3193
final_df.loc[(final_df['city_ascii']== 'Hong Kong'), 'lng'] = 114.1694
final_df.loc[(final_df['city_ascii']== 'Dubai'), 'lat'] = 25.2048
final_df.loc[(final_df['city_ascii']== 'Dubai'), 'lng'] = 55.2708
final_df.loc[(final_df['city_ascii']== 'Prague'), 'lat'] = 50.0755
final_df.loc[(final_df['city_ascii']== 'Prague'), 'lng'] = 14.4378
print(final_df.shape)
```

```python
final_df.to_csv('all-data-best-city.csv')
```
